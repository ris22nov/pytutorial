### 1. Group By

```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data_3.csv')
df
```

```python
df_city = df.groupby('city')
for city,city_df in df_city:
    print(city)
    print(city_df)
```

```python
for city,city_df in df_city:
    max_temp = city_df['temp'].max()
    print(f'Max Temp in {city} is {max_temp}')
```

```python
for city,city_df in df_city:
    avg_ws = city_df['windspeed'].mean()
    print(f'Average Wind Speed in {city} is {avg_ws}')
```

### 2. Concat

```python
india_weather = pd.DataFrame({
    'City': ['Mumbai','Delhi','Pune','Hyderabad'],
    'Temp': [32,26,30,34],
    'Wind Speed':[6,8,3,1]
})

us_weather = pd.DataFrame({
    'City': ['New York','California','Los Angeles'],
    'Temp': [20,12,24],
    'Wind Speed':[5,8,12]
})

weather = pd.concat([india_weather,us_weather],keys=['India','US'])
weather
```

### 3. Merge

```python
df1 = pd.DataFrame({
    'City': ['Mumbai','Delhi','Pune','Hyderabad'],
    'Temp': [32,26,30,34]
})

df2 = pd.DataFrame({
    'City': ['Mumbai','Delhi','Pune','Hyderabad'],
    'Wind Speed':[6,8,3,1]
})

df= pd.merge(df1,df2,on='City')
df
```
