```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
data = pd.read_csv('data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X,Y)
plt.show()
```

```python
X_Mean = np.mean(X)
Y_Mean = np.mean(Y)

num = 0
den = 0

for i in range(len(X)):
    num += (X[i] - X_Mean)*(Y[i]-Y_Mean)
    den += (X[i] - X_Mean)**2
    
m = num/den
c = Y_Mean - (m*X_Mean)

print(m)
print(c)
```

```python
Y_pred = m*X + c

plt.scatter(X, Y) # actual
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
