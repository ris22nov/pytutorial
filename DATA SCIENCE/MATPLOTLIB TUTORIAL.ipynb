## Matplotlib Tutorial

### 1. Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,50,45,43]
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x,y,color='red',linewidth = 4,marker='D',linestyle='--',alpha=0.2)
```

```python
x = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x,max_t,label='Max')
plt.plot(x,min_t,label='Min')
plt.plot(x,avg_t,Label='Avg')

plt.legend(shadow=True)

plt.grid()
```

### 2. Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['Google','Amazon','Microsoft','Facebook']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
plt.ylabel('Revenue')
plt.bar(company,revenue,label='Revenue')
plt.bar(company,profit,label='Profit')
plt.legend()
```

```python
xpos = np.arange(len(company))
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.xlabel('Revenue')
plt.barh((xpos-0.2),revenue,height=0.4,label='Revenue')
plt.barh((xpos+0.2),profit,height=0.4,label='Profit')
plt.legend()
```

### 3. Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar)
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95)
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.title('Blood Sugar Analysis')
plt.xlabel('Sugar Level')
plt.ylabel('No of Patients')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,label=['Men','Women'])
plt.legend()
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.title('Blood Sugar Analysis')
plt.ylabel('Sugar Level')
plt.xlabel('No of Patients')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,label=['Men','Women'],orientation='horizontal')
plt.legend()
```

### 4. PIE CHART

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exps_label = ['Home Rent','Food','Phone/Internet Bill','Car','Other']
exps_value = [1400,600,300,410,250]
```

```python
plt.pie(exps_value,labels=exps_label)
plt.show()
```

```python
plt.pie(exps_value,labels=exps_label,radius=1.5,autopct='%0.2f%%')
plt.show()
```

```python
plt.pie(exps_value,labels=exps_label,radius=1.5,autopct='%0.2f%%',explode=[0,0.2,0,0,0])
plt.show()
```

### 5. Scatter Plot
