```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Bengaluru_House_Data.csv')
df.head()
```

```python
df1 = df.drop(['area_type','availability','society','balcony'],axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2 = df1.replace({
    'bath': np.NaN},
    1
)
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
```

```python
df3.isnull().sum()
```

```python
df3['size'].unique()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.drop('size',axis=1)
```

```python
df3['total_sqft'].unique()
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
            return (float(float(tokens[0])+float(tokens[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
df4['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
```

```python
df4
```

```python
df4.isnull().sum()
```

```python
df5 = df4.dropna()
```

```python
df6 = df5.copy()
```

```python
df6['price_per_sqft'] = (df6['price']*100000)/df6['total_sqft']
```

```python
df6.head()
```

```python
len(df6['location'].unique())
```

```python
df6['location'] = df6['location'].apply(lambda x : x.strip())
```

```python
len(df6['location'].unique())
```

```python
location_stats = df6.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

```python
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
```

```python
location_stats_less_than_10
```

```python
df6['location'] = df6['location'].apply(lambda x : 'Other' if x in location_stats_less_than_10 else x)
```

```python
len(df6['location'].unique())
```

```python
df6[df6['total_sqft']/df6['bhk']<300].head()
```

```python
df6 = df6[~(df6['total_sqft']/df6['bhk']<300)]
```

```python
len(df6['location'].unique())
```

```python
df6['price_per_sqft'].describe()
```

```python
df6.shape
```

```python
def remove_pps_outlier(df):
    df_out = pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m = np.mean(subdf['price_per_sqft'])
        sd = np.std(subdf['price_per_sqft'])
        reduced_df = subdf[(subdf['price_per_sqft'] > (m-sd)) & (subdf['price_per_sqft'] <= (m+sd))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
```

```python
df7 = remove_pps_outlier(df6)
```

```python
df7.shape
```

```python
df7
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

```python
df9 = df8[df8.bath<df8.bhk+2]
df9.shape
```

```python
df9.head()
```

```python
df10 = df9.drop(['size','price_per_sqft'],axis='columns')
df10.head()
```

```python
dummies = pd.get_dummies(df10['location'])
dummies.head()
```

```python
df11 = pd.concat([df10,dummies.drop('Other',axis='columns')],axis='columns')
df11
```

```python
df12 = df11.drop('location',axis='columns')
df12
```

### MODEL BUILDING

```python
df12.shape
```

```python
X = df12.drop(['price'],axis='columns')
X.head()
```

```python
Y = df12['price']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,Y_train)
lr_clf.score(X_test,Y_test)
```

```python
def predict_price(location,sqft,bath,bhk):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar',1000, 2, 2)
```

```python
predict_price('1st Phase JP Nagar',1000, 3, 3)
```

```python
predict_price('Indira Nagar',1000, 2, 2)
```

```python
predict_price('Indira Nagar',2000,3,3)
```
