## Distribution Plot

1. Dist Plot
2. Join Plot
3. Pair Plot

```python
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
```

```python
df.head()
```

### Correlation with Heatmap

A correlation heatmap uses coloured cells, typically in monochromatic scale, to show a 2D correlation matrix between two dicreet dimension or event types.It is very important in feature selection. Only possble with FLoat or Integer data types

### Analysis of TIps Dataframe

**Dependent Feature** - tip  
**Independent Feature** - total_bil,sex,smoker,day,time,size

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### JoinPlot

A Joinplot allows to study relationship between 2 numeric variable.The Central Chart display their correlation.it is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

### Bivariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

### Pair Plot

A 'Pairs plot' is also known as a scatterplot, in which one variable in same data row is matched with another variable value.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

### Dist Plot

Dist Plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Feature

1. Box Plot
2. Violin Plot
3. Count Plot
4. Bar Plot

```python
import seaborn as sns
df =sns.load_dataset('tips')
df.head()
```

**Categorical Feature** - sex,smoker,day,time

### Count Plot

```python
sns.countplot('sex',data=df)
```

### Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='day',y='tip',data=df)
```

```python
sns.barplot(x='day',y='total_bill',data=df)
```

### Box Plot

```python
sns.boxplot('sex','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',data=df)
```

```python
sns.boxplot('day','total_bill',hue='sex',data=df)
```

```python
sns.boxplot(data=df,orient='v')
```

### Violin Plot

Violin Plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
